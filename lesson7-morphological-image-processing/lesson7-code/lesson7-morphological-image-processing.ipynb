## 二值腐蚀

```python
import cv2
import matplotlib.pyplot as plt


def binary_erosion(image_path, kernel_size):
    # 读取图像
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 定义腐蚀核
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))

    # 进行二值腐蚀操作
    eroded_image = cv2.erode(image, kernel)

    return image, eroded_image


image_path = "8-1.png"  # 替换为实际的图像路径
kernel_size = 5  # 可根据需要调整腐蚀核大小

original_image, eroded_image = binary_erosion(image_path, kernel_size)

# 绘制腐蚀前的图片
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

# 绘制腐蚀后的图片
plt.subplot(1, 2, 2)
plt.imshow(eroded_image, cmap='gray')
plt.title("Eroded Image")

plt.show()
```

## 二值膨胀

```python
import cv2
import matplotlib.pyplot as plt

def binary_dilation(image_path, kernel_size):
    # 读取图像
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 定义膨胀核
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))

    # 进行二值膨胀操作
    dilated_image = cv2.dilate(image, kernel)

    return image, dilated_image


image_path = "8-1.png"  # 替换为实际的图像路径
kernel_size = 5  # 可根据需要调整膨胀核大小

original_image, dilated_image = binary_dilation(image_path, kernel_size)

# 绘制膨胀前的图片
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

# 绘制膨胀后的图片
plt.subplot(1, 2, 2)
plt.imshow(dilated_image, cmap='gray')
plt.title("Dilated Image")

plt.show()
```

# 开运算

```python
import cv2
import matplotlib.pyplot as plt

def opening_operation(image_path, kernel_size):
    # 读取图像
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 定义结构元素（这里采用矩形结构元素，你也可根据需求换为其他形状）
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))

    # 进行开运算操作
    opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

    return image, opened_image

image_path = "8-1.png"  # 替换为实际的图像路径
kernel_size = 5  # 可根据需要调整结构元素大小

original_image, opened_image = opening_operation(image_path, kernel_size)

# 绘制操作前的图片
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

# 绘制操作后的图片
plt.subplot(1, 2, 2)
plt.imshow(opened_image, cmap='gray')
plt.title("Opened Image")

plt.show()
```

# 闭运算

```python
import cv2
import matplotlib.pyplot as plt

def closing_operation(image_path, kernel_size):
    # 读取图像
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 定义结构元素（这里采用矩形结构元素，你也可根据需求换为其他形状）
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))

    # 进行闭运算操作
    closed_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

    return image, closed_image

image_path = "8-1.png"  # 替换为实际的图像路径
kernel_size = 5  # 可根据需要调整结构元素大小

original_image, closed_image = closing_operation(image_path, kernel_size)

# 绘制操作前的图片
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

# 绘制操作后的图片
plt.subplot(1, 2, 2)
plt.imshow(closed_image, cmap='gray')
plt.title("Closed Image")

plt.show()
```

# 击中击不中变换

```python
import cv2
import matplotlib.pyplot as plt

def hit_or_miss(image_path, kernel_erosion, kernel_dilation):
    # 读取图像
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 定义腐蚀核与膨胀核
    kernel_erosion = cv2.getStructuringElement(cv2.MORPH_RECT, kernel_erosion)
    kernel_dilation = cv2.getStructuringElement(cv2.MORPH_RECT, kernel_dilation)

    # 进行击中击不中变换
    hitmiss_image = cv2.morphologyEx(image, cv2.MORPH_HITMISS, kernel_erosion, kernel_dilation)

    return image, hitmiss_image

image_path = "8-1.png"  
kernel_erosion_size = (3, 3)  # 腐蚀核大小，可根据需要调整
kernel_dilation_size = (3, 3)  # 膨胀核大小，可根据需要调整

original_image, hitmiss_image = hit_or_miss(image_path, kernel_erosion_size, kernel_dilation_size)

# 绘制操作前的图片
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

# 绘制操作后的图片
plt.subplot(1, 2, 2)
plt.imshow(hitmiss_image, cmap='gray')
plt.title("Hit-or-Miss Image")

plt.show()
```
